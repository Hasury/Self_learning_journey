### Binning DNA sequences

Write a program which creates nine new folders – one for sequences between 100 and 199 bases long, one for sequences between 200 and 299 bases long, etc. Write out each DNA sequence in the input files to a separate file in the appropriate folder.

```python
for file in os.listdir():
  if file.endswith('.dna'):
    with open(file) as input:
      content=input.read()
      content=content.rstrip('\n')
      if len(content) >= 100 and len(content) < 200:
        target_dir='/content/100-200'
      elif len(content) >= 200 and len(content) < 300:
        target_dir='/content/200-300'
      elif len(content) >= 300 and len(content) < 400:
        target_dir='/content/300-400'
      elif len(content) >= 400 and len(content) < 500:
        target_dir='/content/400-500'
      elif len(content) >= 500 and len(content) < 600:
        target_dir='/content/500-600'
      elif len(content) >= 600 and len(content) < 700:
        target_dir='/content/600-700'
      elif len(content) >= 700 and len(content) < 800:
        target_dir='/content/700-800'
      elif len(content) >= 800 and len(content) < 900:
        target_dir='/content/800-900'
      elif len(content) >= 900 and len(content) < 1000:
        target_dir='/content/900-1000'
      else:
        target_dir='/content/1000+'
      if not os.path.exists(target_dir):
        os.makedirs(target_dir)
      shutil.move(file, os.path.join(target_dir, file))
```

```python
for bin_lower in range(100,1000,100):
  bin_upper=bin_lower+99
  bin_name=str(bin_lower)+'_'+str(bin_upper)
  target_dir='/content/'+bin_name
  if not os.path.exists(target_dir):
    os.makedirs(target_dir)

seq_number=1

for file in os.listdir():
  if file.endswith('.dna'):
    print('reading sequences from {}'.format(file))
    with open(file) as input:
      for line in input:
        dna=line.rstrip('\n')
        length=len(dna)
        print('sequence length is {}'.format(line))
        for bin_lower in range(100,1000,100):
          bin_upper=bin_lower+99
          if length >= bin_lower and length < bin_upper:
            bin_name=str(bin_lower)+'_'+str(bin_upper)
            target_dir='/content/'+bin_name
            output_path=os.path.join(target_dir, 'bin_'+str(seq_number)+'_length_'+str(length)+'.dna')
            output=open(output_path, 'w')
            output.write(dna)
            output.close()
            seq_number+=1
```

### Kmer counting

Write a program that will calculate the number of all kmers of a given length across all DNA sequences in the input files and display just the ones that occur more than a given number of times. You program should take two command line arguments – the kmer length, and the cutoff number.

```python
# for Jupyter notebook
import os
import sys

# Simulate command-line arguments for Jupyter Notebook
sys.argv = ["script_name", "3", "900"]  # Example: kmer_size=3, count_cutoff=2

# Convert command-line arguments to variables
kmer_size = int(sys.argv[1])
count_cutoff = int(sys.argv[2])

# Define the function to split DNA into k-mers
def split_dna(dna, kmer_size):
    kmers = []
    for start in range(len(dna) - (kmer_size - 1)):
        kmer = dna[start:start + kmer_size]
        kmers.append(kmer)
    return kmers

# Create an empty dictionary to hold the counts
kmer_counts = {}

# Process each file with the ".dna" extension
for file_name in os.listdir("."):
    if file_name.endswith(".dna"):
        with open(file_name, "r") as dna_file:  # Use context manager for file handling
            # Process each DNA sequence in the file
            for line in dna_file:
                dna = line.strip()  # Remove newline and extra spaces
                if dna:  # Skip empty lines
                    # Increase the count for each k-mer
                    for kmer in split_dna(dna, kmer_size):
                        kmer_counts[kmer] = kmer_counts.get(kmer, 0) + 1

# Print k-mers whose counts are above the cutoff
for kmer, count in kmer_counts.items():
    if count > count_cutoff:
        print(kmer + " : " + str(count))
```

```python
# for python in the system
import os
import sys

# convert command line arguments to variables
kmer_size = int(sys.argv[1])
count_cutoff = int(sys.argv[2])

# define the function to split dna
def split_dna(dna, kmer_size):
    kmers = []
    for start in range(0,len(dna)-(kmer_size-1),1):
        kmer = dna[start:start+kmer_size]
        kmers.append(kmer)
    return kmers

# create an empty dictionary to hold the counts
kmer_counts = {}

# process each file with the right name
for file_name in os.listdir("."):
    if file_name.endswith(".dna"):
        dna_file = open(file_name)

        # process each DNA sequence in a file
        for line in dna_file:
            dna = line.rstrip("\n")

            # increase the count for each k-mer that we find
            for kmer in split_dna(dna, kmer_size):
                current_count = kmer_counts.get(kmer, 0)
                new_count = current_count + 1
                kmer_counts[kmer] = new_count

# print k-mers whose counts are above the cutoff
for kmer, count in kmer_counts.items():
    if count > count_cutoff:
        print(kmer + " : " + str(count))
```
