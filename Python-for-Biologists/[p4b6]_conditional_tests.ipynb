### Several species

Print out the gene names for all genes belonging to _Drosophila melanogaster_ or _Drosophila simulans_.

```python
for x in range(len(data)):
  if data.loc[x][0]=='Drosophila melanogaster' or data.loc[x][0]=='Drosophila simulans':
    print(data.loc[x][2])
```

### Length range


Print out the gene names for all genes between 90 and 110 bases long.


```python
for x in range(len(data)):
  if len(data.loc[x][1])>90 and len(data.loc[x][1])<110:
    print(data.loc[x][2])
```

### AT content

Print out the gene names for all genes whose AT content is less than 0.5 and whose expression level is greater than 200.

```python
data[3]=data[3].astype(int)
```

```python
for x in range(len(data)):
  seq=data.loc[x][1]
  count=0
  for base in seq:
    if base=='a' or base=='t':
      count+=1
  if count/len(seq)<0.5 and data.loc[x][3]>200:
    print(data.loc[x][2])
```

### Complex condition

Print out the gene names for all genes whose name begins with "k" or "h" except those belonging to _Drosophila melanogaster_.

```python
for x in range(len(data)):
  if data.loc[x][2].startswith('k') or data.loc[x][2].startswith('h') and data.loc[x][0]!='Drosophila melanogaster':
    print(data.loc[x][2])
```

### High low medium

For each gene, print out a message giving the gene name and saying whether its AT content is high (greater than 0.65), low (less than 0.45) or medium (between 0.45 and 0.65).

```python
for x in range(len(data)):
  seq=data.loc[x][1]
  count=0
  for base in seq:
    if base=='a' or base=='t':
      count+=1
  if count/len(seq)>0.65:
    print(data.loc[x][2],'has high AT content')
  elif count/len(seq)<0.45:
    print(data.loc[x][2],'has low AT content')
  else:
    print(data.loc[x][2],'has medium AT content')
```
